```python
def rec_DOM_tree(tree, soup, i):
    parent_id = i
    count = 0
    while soup is not None:
        if soup.name is not None:
            tree.create_node(soup.name, i+count, parent_id-1)
            if soup.next_element != soup.next_sibling:
                (tree, extra) = rec_DOM_tree(tree, soup.next_element, i+count+1)
                count+=extra
        else:
            tree.create_node(soup, i+count, parent_id-1)
        soup = soup.next_sibling
        count+=1
    return (tree, count)
```
